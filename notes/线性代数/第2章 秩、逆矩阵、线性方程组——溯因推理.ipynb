# 第2张 秩、逆矩阵、线性方程组——溯因推理

## 2.1 问题设定：逆问题

由已知结果$y$去推测原因$x$的问题，称为**逆问题**。相对于逆问题，如果结果之间的对照更重要，我们会考虑从原因$x$出发去预测结果$y$，这样的问题称为**顺问题**。

所谓逆问题，其实就是**线性方程组**。

## 2.2 良性问题（可逆矩阵）

### 2.2.1 可逆性与逆矩阵

存在逆矩阵的的方阵$A$，称为**正则矩阵**(可逆矩阵、非奇异矩阵))。而不是正则矩阵的，则称为**奇异矩阵**。

### 2.2.2 线性方程组的解法（系数矩阵可逆的情况）

用分块矩阵表示线性方程组的求解过程：
$$
(A|y)=\Big({x\over -1}\Big)=o
$$
变形，得到
$$
(I|s)\Big({x\over -1}\Big)=o
$$
的形式，则$x-s=o$，$s$的各个分量就是方程中对应变量的解。

更简单的有Gauss-Jordan法。当使用Gauss-Jordan法时，对角元素可能会变成0，这个时候可以采用**选主元**法——互换两行。

### 2.2.3 逆矩阵的计算方法

使用逆矩阵计算公式太麻烦，不推荐使用。

利用
$$
(A|I)\to(I|X) 
$$
的变形，便可得到$A$的逆向矩阵$X$

### 2.2.4 初等变换

笔算法的步骤可以总结为，对形如$(A|y)$或$(A|I)$的矩阵进行如下操作：

- 将某行乘以($c\ne 0$)
- 将某行的$c$倍加到另一行上
- 交换两行

笔算法中的所有操作，都可以看做是在一次左乘形如$Q_i(c)$，$R_{i,j}(c)$，$S_{i,j}$的特殊方阵。这些操作我们成为**初等(行)变换**。

## 2.3 恶性问题

### 2.3.1 恶性问题举例

当原因$x=(x_1,\dots,x_n)^T$和结果$y=(y_1,\dots,y_m)^T$两者维数不同时$n\ne m$，比较麻烦。

对于给定的$A$，在映射的作用下，满足$Ax=o$的$x$的集合称为$A$的**核(kernel)**，记为$Ker A$。

$Ker A$也表示了“压缩扁平化”的变换方向。于是，对于与$Ker A$平行的各元素，在只知道映射后的点的情况下，是无法进行还原的。

对于给定的$A$，将$x$进行各种不同的变换，在$A$的作用下，$y=Ax$构成的集合，称为$A$的像(image)，计做$Im A$。像就是把原空间通过$A$变换到目标空间中时对应的领域。

### 2.3.2 问题的恶劣程度——核与像

对于相同的结果$y$，引起它的原因$x$如果是唯一的，那么映射$y=Ax$是**单射**。对于无论什么样的结果$y$，都可以找到导致它的相应的原因$x$，映射$y=Ax$是**满射**。对于满足上述两者的$y=Ax$称为**双射**。

- $Ker A$仅包含原点$o\Leftrightarrow$映射为单射
- $Im A$与目标空间全体(值域)一致$\Leftrightarrow$映射为满射

### 2.3.3 维数定理

维数定理（秩——零化度定理）:

对于$m\times n$矩阵$A$，有
$$
dim Ker A + dim Im A = n
$$
其中$dim X$表示$X$的维数。

- 若$m<n$($A$是矮矩阵)，则$A$不会是单射
- 若$m>n$($A$是长矩阵)，则$A$不会是满射
